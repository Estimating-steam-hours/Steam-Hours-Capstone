```python
import requests
import pandas as pd
import json
```

```python
response = requests.get('https://steamspy.com/api.php?request=top100in2weeks')
```

```python
response_data = response.json()
```

```python
response_data
```

```python
steamspy_df = pd.DataFrame(response_data).T
```

```python
steamspy_df
```

```python
steamspy_df.to_csv('steamspy_top100_2weeks.csv', index=False)
```
